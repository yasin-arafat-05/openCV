# Probabilistic Hough Transform

`it gives us more better result than hough line transformation`

```python
import cv2 as cv 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
img = cv.imread("data/sudoku.png")

gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
edge = cv.Canny(img,50,150)

lines = cv.HoughLinesP(edge,1,np.pi/180,100,minLineLength=100)


# lines give me (r,theta) then we make  (r cos (theta) + r sin (theta))
print(f"line: {lines}")

for line in lines:
   x1,y1,x2,y2 = line[0]
   cv.line(img,(x1,y1),(x2,y2),(0,0,255),2)


imges = [img,gray,edge]
titles = ["orginal","Gray","Canny"]

plt.figure(figsize=(12,12))
for i in range(3):
    plt.subplot(1,3,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.title(titles[i])
    plt.imshow(imges[i])
if cv.waitKey(0) == ord('q'):
    cv.destroyAllWindows()
    

    
```
