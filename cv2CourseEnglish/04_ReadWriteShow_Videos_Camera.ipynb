```python
import cv2 as cv
```

```python
# from camera:

cap = cv.VideoCapture(0)

while True: # cap.isOpened() give us boolean value
    rate,frame = cap.read() # boolen, img
    cv.imshow("output",frame)
    if not rate:
        break
    if cv.waitKey(1) & 0xFF == ord('q'):
        break
cap.release()
cv.destroyAllWindows()
```

# why 0 ? 

- `0`: Represents the index of the first camera device connected to the system.
- `1`, `2`, `3`, etc.: Represent the indices of subsequent camera devices if there are multiple cameras connected.

# why we use cap.release()

In OpenCV, `cap.release()` is used to release the resources associated with the `VideoCapture` object. When you're done using a video capture device (like a webcam), it's important to release it properly to free up system resources and allow other applications to access the camera.

Here's why `cap.release()` is important:

1. **Resource Management**: When you open a video capture device using `cv2.VideoCapture()`, it allocates resources such as memory and file descriptors to access the camera. If you don't release these resources properly, it can lead to memory leaks or cause other applications to fail when trying to access the camera.

2. **Clean Exit**: Releasing the video capture device ensures a clean exit from your program. It's good practice to release resources when they're no longer needed to avoid potential issues and maintain system stability.

3. **Allowing Other Applications**: Releasing the camera allows other applications to access it. If your program holds onto the camera without releasing it, other applications won't be able to use the camera until yours exits or releases the camera resource.

# Open an show a video from file

```python
# from video file:

cap = cv.VideoCapture('data/vtest.avi') # give file path here

while cap.isOpened(): 
    rate,frame = cap.read() 
    if not rate:
        break
    
    cv.imshow("output",frame)
    
    if cv.waitKey(60) & 0xFF == ord('q'): # wait key(val) value work as frame per second
        break
cap.release()
cv.destroyAllWindows()
```

# Open camera in gray mode:

- grayFrame = cv.cvtColor(frame,cv.COLOR_BGR2GRAY)
- openCV read in BGR
- that why we convert BGR to Gray not RGB 
    

```python
# from video file:

cap = cv.VideoCapture('data/vtest.avi') # give file path here

while cap.isOpened(): 
    rate,frame = cap.read() 
    if not rate:
        break
    
    grayFrame = cv.cvtColor(frame,cv.COLOR_BGR2GRAY)
    
    cv.imshow("orginal",frame)
    cv.imshow("grayFrame",grayFrame)
    
    if cv.waitKey(60) & 0xFF == ord('q'): # wait key(val) value work as frame per second
        break
cap.release()
cv.destroyAllWindows()
```

# Read video orginal height and width:

- more: go to documentation then videoio then flags for video I/O

```python
# from video file:

cap = cv.VideoCapture('data/vtest.avi') # give file path here



while cap.isOpened(): 
    rate,frame = cap.read() 
    if not rate:
        break
    cv.imshow("video",frame)
    if cv.waitKey(60) & 0xFF == ord('q'): # wait key(val) value work as frame per second
        break
    print("height: {} ".format(cap.get(cv.CAP_PROP_FRAME_HEIGHT)))
    print("width: {} ".format(cap.get(cv.CAP_PROP_FRAME_WIDTH)))
    
cap.release()
cv.destroyAllWindows()
```
