# Lecture 13: How to make TrackBar in OPENCV:

`It's usefull when we try to change the value of pixel in runtime.`

<br>


```python
import cv2 as cv 
import numpy as np 
```



# Example ONE: 


```python
img = np.zeros((300,512,3),np.uint8)


# call back function for onChange
def onChange(x):
    print(x)

# crate a window with name:
cv.namedWindow("image")

cv.createTrackbar('B','image',0,255,onChange)
cv.createTrackbar('G','image',0,255,onChange)
cv.createTrackbar('R','image',0,255,onChange)

while True:
    
    cv.imshow("image",img) # show in the same window that we make begining of the code.
    
    if cv.waitKey(1) & 0xFF == ord('q'):
       break
   
    # get the tracker value
    b = cv.getTrackbarPos('B',"image")
    g = cv.getTrackbarPos('G',"image")
    r = cv.getTrackbarPos('R',"image")
    
    img[:] = [b,g,r]
    


cv.destroyAllWindows()

```


# Example Two  (make gray imge): 


```python
import cv2 as cv

img = cv.imread('data/lena.jpg')

# callback function for onChange
def onChange(x):
    pass

# create a window with name
cv.namedWindow("image")

# create tracker for switch and value
cv.createTrackbar('S', 'image', 0, 1, onChange)
cv.createTrackbar('CP', 'image', 20, 300, onChange)

while True:
    # Create a copy of the original image
    img_copy = img.copy()

    if cv.waitKey(1) & 0xFF == ord('q'):
        break

    # get the tracker value
    s = cv.getTrackbarPos('S', 'image')
    text = str(cv.getTrackbarPos('CP', 'image'))
    cv.putText(img_copy, text, (50, 50), cv.FONT_HERSHEY_SIMPLEX, 1, (255, 0, 0))
    
    if s != 0:
        img_copy = cv.cvtColor(img_copy, cv.COLOR_BGR2GRAY)
    
    cv.imshow("image", img_copy)

cv.destroyAllWindows()
```
