- `Smoothing or Blurring `



```python
import cv2 as cv 
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
img = cv.imread("data/opencv-logo-white.png")


img = cv.cvtColor(img,cv.COLOR_BGR2RGB)

kernel = (1/25) * np.ones((5,5))  

# homogeneous filter
dst = cv.filter2D(img,-1,kernel)

# use blur method
blur = cv.blur(img,(5,5)) 

# use gaussian filter:
gaussian = cv.GaussianBlur(img,(5,5),0)


# median blur filter kernal size should be odd
# usefull working with solt and paper problem

median = cv.medianBlur(img,5)

# bileratel filter (Removing filte, while keeping the edge sharp)
bileratel = cv.bilateralFilter(img,9,75,75)

images = [img,dst,blur,gaussian,median,bileratel]
titles = ["orginalImg","2D_Conv","blur","gaussian","median","bileteral"]


plt.figure(figsize=(10, 10))

for i in range(6):
    plt.subplot(2,3,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.title(titles[i])
    plt.imshow(images[i],'gray')

plt.show()
    

if cv.waitKey(0) == ord('q'):
    cv.destroyAllWindows()
```
