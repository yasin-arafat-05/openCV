`Matplotlib(Ploting 2D library in python ) with openCV`

```python
import cv2 as cv 
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
img = cv.imread("data/lena.jpg")

cv.imshow("openCV",img)

print("Matplotlib Image: in BGR ")
plt.xticks([]) # hide xticks in x axis is gone 
plt.yticks([]) # hide yticks in y axis is gone 

plt.imshow(img) # read the image in BGR
if cv.waitKey(0) == ord('q'):
    cv.destroyAllWindows()
    
```

# show multiple image in one frame: 

```python
orginal = cv.imread('data/gradient.png')

_,th1 = cv.threshold(orginal,50,255,cv.THRESH_BINARY) # (src,ThesholdValue,maximumThesholdValue,ThesholdType)
_,th2 = cv.threshold(orginal,200,255,cv.THRESH_BINARY_INV) 
_,th3 = cv.threshold(orginal,127,255,cv.THRESH_TRUNC)
_,th4 = cv.threshold(orginal,127,255,cv.THRESH_TOZERO)
_,th5 = cv.threshold(orginal,127,255,cv.THRESH_TOZERO_INV)

titles = ["ORGINAL","BINARY","BINARY_INV","TRUNC","TOZERO","TOZERO_INV"]
images = [orginal,th1,th2,th3,th4,th5]


for i in range(6):
    plt.subplot(2,3,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.xlabel(titles[i])
    plt.imshow(images[i])
    
```
