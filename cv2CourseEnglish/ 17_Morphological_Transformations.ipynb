`Morphological transformations are some simple operations based on the image shape now morphological transformation is normally performed on a **binary image** and when we perform morphological transformation there are **two things** which are required first is the original image and second is called a structuring element or a kernel which decides the nature of the operation.`

```python
import numpy as np 
import cv2 as cv 
import matplotlib.pyplot as plt 
```

```python

img = cv.imread('data/smarties.png',0) # read the image in grayscale


# we performed morphological transformations in binary images : we need a mask

_,mask = cv.threshold(img,200,255,cv.THRESH_BINARY_INV)


# in mask there are some black plot. we replace this is white color by dialation
kernal = np.ones((5,5),np.uint8)
dilation = cv.dilate(mask,kernal,iterations=2)

# After dilation the size of our ball increased : eroson is inverse of dilation
eroson = cv.erode(mask,kernal,iterations=2)

# opening = erosion + dialation:
opening = cv.morphologyEx(mask,cv.MORPH_OPEN,kernal)

# closing = dilation + erosion:
closing = cv.morphologyEx(mask,cv.MORPH_CLOSE,kernal)
titles = ["image","mask","dilation_img","eroson","opening","closing"]
images = [img,mask,dilation,eroson,opening,closing]



for i in range(6):
    plt.subplot(3,2,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.title(titles[i])
    plt.imshow(images[i],'gray')
    
    
```
