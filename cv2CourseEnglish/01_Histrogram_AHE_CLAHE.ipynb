```python
import cv2 as cv 
import matplotlib.pyplot as plt 
```

```python
aero1 = cv.imread("cv2CourseEnglish/data/aero1.jpg")
aero3 = cv.imread("cv2CourseEnglish/data/aero3.jpg")
busketball = cv.imread("cv2CourseEnglish/data/basketball1.png")
lena = cv.imread("cv2CourseEnglish/data/lena.jpg")
```

# Make histrogram:

<br>


```python
gray = cv.cvtColor(busketball,cv.COLOR_BGR2GRAY)
# we need pass in list. (source,channel(gary is one we gave 0 for that),mask,histSize,range)
# mask: If we give mask that pixel is not consider in the histrogram
# histSize: total number of bins (256)
# range: (0~255)
his = cv.calcHist([gray],[0],None,[256],[0,255])
plt.figure(figsize=(8,8))
plt.plot(his) # for ploting histrogram we use plot of matplotlib.
```

```python

histoEqu = cv.equalizeHist(gray)

fig,axis = plt.subplots(nrows=2,ncols=2,figure=(10,50))
fig.tight_layout()


axis[0][0].imshow(histoEqu)
axis[0][0].set_title("histrogram equalization")
axis[0][0].axis("off")


axis[0][1].imshow(busketball)
axis[0][1].set_title("Real Image")
axis[0][1].axis("off")


hiseq = cv.calcHist([histoEqu],[0],None,[256],[0,255])
axis[1][0].plot(hiseq)
axis[1][0].set_title("histogram of equilize image")



axis[1][1].plot(his)
axis[1][1].set_title("histogram of Real Image")


```
