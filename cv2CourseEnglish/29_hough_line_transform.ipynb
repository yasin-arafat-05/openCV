```python
import cv2 as cv
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
concept = cv.imread("photo/image-7.png")
plt.figure(figsize=(10,10))
plt.imshow(concept)
```

```python
img = cv.imread("data/sudoku.png")

gray = cv.cvtColor(img,cv.COLOR_BGR2GRAY)
edge = cv.Canny(img,50,150)

lines = cv.HoughLines(edge,1,np.pi/180,200)


# lines give me (r,theta) then we make  (r cos (theta) + r sin (theta))
print(f"line: {lines}")

for line in lines:
    rho, theta = line[0]
    a = np.cos(theta)
    b = np.sin(theta)
    x0 = a * rho 
    y0 = b * rho 
    x1 = int(x0 - 1000 * b)
    y1 = int(y0 + 1000 * a)
    x2 = int(x0 + 1000 * b)
    y2 = int(y0 - 1000 * a)
    cv.line(img,(x1,y1),(x2,y2),(0,0,255),2)


imges = [img,gray,edge]
titles = ["orginal","Gray","Canny"]

plt.figure(figsize=(12,12))
for i in range(3):
    plt.subplot(1,3,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.title(titles[i])
    plt.imshow(imges[i])
if cv.waitKey(0) == ord('q'):
    cv.destroyAllWindows()
    

    
```
