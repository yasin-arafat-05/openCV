```python
import cv2 as cv 
```

```python
# read and show image:

img = cv.imread('./picture/image.png')
img  = cv.resize(img,(600,480)) 
cv.imshow("win",img)
if cv.waitKey(0) == ord('q'):
    cv.destroyAllWindows()
    
```

```python
# read and show video and its height and width:

cap = cv.VideoCapture('./picture/BMW.mp4')
while True:
    rate,frame = cap.read() 
    print(f"{cap.get(cv.CAP_PROP_FRAME_HEIGHT)}") # height of the orginal video
    print(f"{cap.get(cv.CAP_PROP_FRAME_WIDTH)}") # width of the orginal video
    if not rate:
        print("video finished")
    frame = cv.resize(frame,(520,480))
    cv.imshow('video',frame)
    if cv.waitKey(1) == ord('q'):
        break
cap.release()
cv.destroyAllWindows()

    
```

```python
# read video from camera

cap = cv.VideoCapture(0)  # as we have only one camera then the id of my camera is 0

while True:
    rate,frame = cap.read()
    # set width and height 
    cap.set(3,640) # for width id is 3
    cap.set(4,480) # for height id is 4
    cap.set(10,100) # set the brightness id 10
    if not rate:
        print("camera is turn off")
    cv.imshow('video',frame)
    if cv.waitKey(1) == ord('q'):
        break
cap.release()
cv.destroyAllWindows()
```
